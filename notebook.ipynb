## Part 2: Optimization: The Hospital Shift Scheduler

### Problem

The goal is to generate a valid weekly hospital shift schedule by assigning nurses to all 21 shifts while satisfying leave-day, rest-period, and maximum workload constraints.

### Approach

This is solved as a Constraint Satisfaction Problem. The variables are weekly shifts, the domains are available nurses, and the constraints are leave days, Night-to-next-Morning rest requirements, and a maximum of 5 shifts per nurse. The solver uses node consistency, AC-3, MRV variable selection, recursive backtracking, and forward checking.

````python
# Display the Part 2 solution code.
print(Path('part2/shift_solver.py').read_text(encoding='utf-8'))
````

### Part 2 Test Cases

The following test cell runs all three staff files. Each result must assign all 21 shifts and satisfy every constraint.

````python
sys.path.append('part2')
from shift_solver import Shift_AI_Solver

def run_shift_test(staff_file):
    solver = Shift_AI_Solver(staff_file)
    print(f'Generating Weekly Schedule for 2026 using {staff_file}...')
    print(f'Data loaded: {len(solver.nurses)} staff members available.')
    print('Enforcing node consistency...')
    solver.enforce_node_consistency()
    print('Running AC-3 algorithm...')
    assert solver.ac3(), 'AC-3 failed: no valid domains remain.'
    print('Starting backtracking search...')
    schedule = solver.backtrack({})
    assert schedule is not None, 'Backtracking failed: no valid schedule found.'
    assert len(schedule) == 21, 'Schedule does not contain all 21 shifts.'
    assert solver.consistent(schedule), 'Final schedule violates a constraint.'
    print(solver.format_schedule(schedule))
    print('\n' + '=' * 70 + '\n')

for staff_file in [
    'part2/staff_small.txt',
    'part2/staff_medium.txt',
    'part2/staff_complex.txt',
]:
    run_shift_test(staff_file)
````
